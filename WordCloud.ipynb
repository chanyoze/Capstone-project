```python
with open('/content/gdrive/MyDrive/캡스톤/crawl.txt', encoding='utf8') as file_object:
    contents = file_object.read()
    #print(contents)
```

```python
#형태소 분석기 Open Korean Text 사용해서 명사만 분리해냄
okt = Okt()
#띄어쓰기 교정 하고 진행함
#spacing = Spacing()

#텍스트 파일 명사만 분리
token = okt.nouns(contents)

#재접합
list = nltk.Text(token)
#new_list= spacing(contents)

print(len(token))
print(len(list))

rank_list = Counter(list).most_common(1000)
print(rank_list)
```

```python
#중복 단어 제거 
remove_same_word = set(list)
#1글자 단어 제거
remove_one_word = [x for x in remove_same_word if len(x)>1]
#불용어 제거
with open('/content/gdrive/MyDrive/캡스톤/stopwords.txt', encoding='utf8') as file_object:
    #list_file = file_object.readlines()
    stopwords = file_object.read().split(',')


remove_stopwords = [x for x in remove_one_word if x not in stopwords]

rank_list = Counter(remove_stopwords).most_common(1000)

print(len(remove_same_word), len(remove_one_word), len(remove_stopwords), len(rank_list))
print(remove_same_word) 
print(remove_one_word)
print(remove_stopwords)
```
